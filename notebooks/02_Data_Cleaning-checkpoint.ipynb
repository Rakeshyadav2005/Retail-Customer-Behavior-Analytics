```python
import pandas as pd
import numpy as np
```

```python
events = pd.read_csv("../data/raw/events.csv")

categories = pd.read_csv("../data/raw/category_tree.csv")

properties_part1 = pd.read_csv("../data/raw/item_properties_part1.csv")

properties_part2 = pd.read_csv("../data/raw/item_properties_part2.csv")
```

```python
print("Events:", events.shape)
print("Categories:", categories.shape)
print("Properties Part 1:", properties_part1.shape)
print("Properties Part 2:", properties_part2.shape)
```

```python
properties = pd.concat(
    [properties_part1, properties_part2],
    ignore_index=True
)
```

```python
# Convert transactionid to nullable integer
events["transactionid"] = (
    pd.to_numeric(events["transactionid"], errors="coerce")
      .astype("Int64")
)
```
