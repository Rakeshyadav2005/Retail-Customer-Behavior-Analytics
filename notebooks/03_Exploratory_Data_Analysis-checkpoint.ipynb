```python
import pandas as pd
```

```python
events = pd.read_csv(
    "../data/cleaned/events_cleaned.csv",
    parse_dates=["timestamp"]
)

categories = pd.read_csv(
    "../data/cleaned/categories_cleaned.csv"
)

properties = pd.read_csv(
    "../data/cleaned/properties_cleaned.csv",
    parse_dates=["timestamp"]
)
```

```python
round(events["event"].value_counts(normalize=True) * 100, 2)
```

```python
views = (events["event"] == "view").sum()
print("Views:", views)
```

```python
cart = (events["event"] == "addtocart").sum()
print("Add to Cart:", cart)
```

```python
transactions = (events["event"] == "transaction").sum()
print("Transactions:", transactions)
```

```python
conversion_rate = (transactions / views) * 100

print(f"Conversion Rate: {conversion_rate:.2f}%")
```

```python
cart_rate = (cart / views) * 100

print(f"Add-to-Cart Rate: {cart_rate:.2f}%")
```

```python
purchase_rate = (transactions / cart) * 100

print(f"Cart to Purchase Rate: {purchase_rate:.2f}%")
```

```python
events_per_visitor = len(events) / events["visitorid"].nunique()

print(f"Average Events per Visitor: {events_per_visitor:.2f}")
```

```python
events["visitorid"].value_counts().head(10)
```

```python
top_viewed = (
    events[events["event"] == "view"]
    .groupby("itemid")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_viewed
```

```python
top_cart = (
    events[events["event"] == "addtocart"]
    .groupby("itemid")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_cart
```

```python
top_purchased = (
    events[events["event"] == "transaction"]
    .groupby("itemid")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_purchased
```

```python
top_viewed = top_viewed.reset_index()
top_viewed.columns = ["Item ID", "Views"]

top_viewed
```

```python
top_cart = top_cart.reset_index()
top_cart.columns = ["Item ID", "Cart Count"]

top_cart
```

```python
top_purchased = top_purchased.reset_index()
top_purchased.columns = ["Item ID", "Purchases"]

top_purchased
```

```python
top_viewed.merge(
    top_purchased,
    on="Item ID",
    how="left"
)
```

```python
item_category = properties[
    properties["property"] == "categoryid"
][["itemid", "value"]]

item_category.head()
```

```python
item_category.columns = ["itemid", "categoryid"]

item_category.head()
```

```python
events_category = events.merge(
    item_category,
    on="itemid",
    how="left"
)

events_category.head()
```

```python
events_category.shape
```

```python
top_categories = (
    events_category[
        events_category["event"] == "view"
    ]
    .groupby("categoryid")
    .size()
    .sort_values(ascending=False)
    .head(10)
)

top_categories
```
